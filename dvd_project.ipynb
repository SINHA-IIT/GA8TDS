```python
# prev_df = pd.read_csv("/content/previous_application.csv")
# prev_df.head()


# Helper functions for missing values

def missing_value_report(df, dataset_name):
    print(f"\n===============================")
    print(f" Missing Value Report: {dataset_name}")
    print(f"===============================")

    missing_count = df.isnull().sum()
    missing_percent = (missing_count / len(df)) * 100

    missing_df = pd.DataFrame({
        "Column": df.columns,
        "Missing Count": missing_count,
        "Missing %": missing_percent.round(2)
    }).sort_values(by="Missing %", ascending=False)

    display(missing_df.head(20))

    # Categorize severity
    missing_df["Severity"] = pd.cut(
        missing_df["Missing %"],
        bins=[-0.1, 5, 20, 50, 100],
        labels=["Low (<5%)", "Moderate (5–20%)", "High (20–50%)", "Severe (>50%)"]
    )

    print("\nSeverity breakdown:")
    display(missing_df["Severity"].value_counts())

    return missing_df

def plot_missing(df, dataset_name):
    missing_percent = (df.isnull().sum() / len(df)) * 100
    missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

    plt.figure(figsize=(10,6))
    missing_percent.head(20).plot(kind="barh")
    plt.title(f"Top 20 Columns with Missing Values – {dataset_name}")
    plt.xlabel("Missing %")
    plt.ylabel("Column Name")
    plt.gca().invert_yaxis()
    plt.show()
```

```python
# print(app_df.shape, prev_df.shape, desc_df.shape)

# Cell 6: Missing value analysis

app_missing = missing_value_report(app_df, "Application Data")
prev_missing = missing_value_report(prev_df, "Previous Application")
desc_missing = missing_value_report(desc_df, "Columns Description")

plot_missing(app_df, "Application Data")
plot_missing(prev_df, "Previous Application")
```

```python
#  Merge a simple feature from previous_application to application_data
#  number of previous applications per customer

prev_counts = prev_df.groupby("SK_ID_CURR")["SK_ID_PREV"].count().reset_index()
prev_counts.rename(columns={"SK_ID_PREV": "PREV_APP_COUNT"}, inplace=True)

app_merged = app_df.merge(prev_counts, on="SK_ID_CURR", how="left")
app_merged["PREV_APP_COUNT"] = app_merged["PREV_APP_COUNT"].fillna(0)

plt.figure(figsize=(8,5))
sns.barplot(
    x="PREV_APP_COUNT",
    y="TARGET",
    data=app_merged[app_merged["PREV_APP_COUNT"] <= 10],  # focus on 0–10 previous apps
    estimator=np.mean
)
plt.title("Default Rate by Number of Previous Applications (<=10)")
plt.xlabel("Number of Previous Applications")
plt.ylabel("Default Rate")
plt.show()
```
